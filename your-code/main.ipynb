```python
numerical_cols = ["Fresh", "Milk", "Grocery", "Frozen", "Delicassen"]
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    
    # Count outliers
    num_outliers = len(outliers)
    
    # Calculate percentage of outliers
    total_rows = len(df)
    percentage_outliers = (num_outliers / total_rows) * 100 if total_rows > 0 else 0

    print(f"Column '{col}': {num_outliers} outliers ({percentage_outliers:.2f}%)")
```

```python
# Define the columns that represent sales/expenditures
sales_cols = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Delicassen']

# Calculate Total Sales for Each Customer
df['Total_Sales'] = df[sales_cols].sum(axis=1)

# Sort Customers by Total Sales (Descending)
df_sorted = df.sort_values(by='Total_Sales', ascending=False).reset_index(drop=True)

# Calculate Cumulative Sums and Percentages
df_sorted['Cumulative_Sales_Pct'] = (df_sorted['Total_Sales'].cumsum() / df_sorted['Total_Sales'].sum()) * 100
df_sorted['Customer_Pct'] = (np.arange(1, len(df_sorted) + 1) / len(df_sorted)) * 100

# Find the "80/20" Point (first row where Cumulative_Sales_Pct is >= 80%)
pareto_idx = df_sorted[df_sorted['Cumulative_Sales_Pct'] >= 80].index[0]
customer_pct_at_80_sales = df_sorted.loc[pareto_idx, 'Customer_Pct']

print(f"Total customers: {len(df_sorted)}")
print(f"Total sales: {df_sorted['Total_Sales'].sum():,.2f}\n")

print("--- Pareto Principle Check (80/20 Rule) ---")
print(f"Approximately {customer_pct_at_80_sales:.2f}% of customers account for 80% of total sales.")

if customer_pct_at_80_sales <= 25:
    print("\nThis dataset appears to largely display the Pareto Principle (or even a stronger concentration).")
else:
    print("\nThis dataset does not strongly display the Pareto Principle (80/20 rule).")

# Optional: Display a few rows around the 80% mark for inspection
# print("\n--- Rows around the 80% sales mark ---")
# print(df_sorted[['Total_Sales', 'Cumulative_Sales_Pct', 'Customer_Pct']].iloc[max(0, pareto_idx - 2):min(len(df_sorted), pareto_idx + 3)])
```

# Challenge 2 - Data Cleaning and Transformation

If your conclusion from the previous challenge is the data need cleaning/transformation, do it in the cells below. However, if your conclusion is the data need not be cleaned or transformed, feel free to skip this challenge. But if you do choose the latter, please provide rationale.

```python
def remove_outliers_iqr_union(df, columns_to_clean, factor=1.5):

    cleaned_df = df.copy()
    
    # Initialize an empty boolean Series, all True (meaning no rows are marked for removal yet)
    # This mask will accumulate all rows that are outliers in *any* of the columns
    outlier_mask_all_columns = pd.Series([False] * len(cleaned_df), index=cleaned_df.index)

    print("Detecting outliers across specified columns:")
    print("-" * 70)

    for col in columns_to_clean:
        if col in cleaned_df.columns and pd.api.types.is_numeric_dtype(cleaned_df[col]):
            if not cleaned_df[col].empty:
                Q1 = cleaned_df[col].quantile(0.25)
                Q3 = cleaned_df[col].quantile(0.75)
                IQR = Q3 - Q1
                
                lower_bound = Q1 - factor * IQR
                upper_bound = Q3 + factor * IQR
                
                # Create a boolean mask for outliers in the current column
                col_outlier_mask = (cleaned_df[col] < lower_bound) | (cleaned_df[col] > upper_bound)
                
                num_outliers_in_col = col_outlier_mask.sum()
                print(f"  Column '{col}': {num_outliers_in_col} outliers detected (IQR bounds: [{lower_bound:.2f}, {upper_bound:.2f}])")
                
                # Combine this column's outliers with the overall outlier mask using OR
                outlier_mask_all_columns = outlier_mask_all_columns | col_outlier_mask
            else:
                print(f"  Column '{col}' is empty or contains only non-numeric data. Skipping outlier detection for this column.")
        else:
            print(f"  Column '{col}' not found or is not a numeric type. Skipping outlier detection for this column.")

    print("-" * 70)
    
    # Filter the DataFrame: keep rows where the overall outlier mask is False
    initial_rows = len(cleaned_df)
    df_without_outliers = cleaned_df[~outlier_mask_all_columns]
    
    removed_rows = initial_rows - len(df_without_outliers)
    print(f"\nTotal rows removed (outlier in at least one specified column): {removed_rows} out of {initial_rows} ({removed_rows/initial_rows*100:.2f}%)")

    return df_without_outliers
```

```python
df_without_outliers = remove_outliers_iqr_union(df, columns_to_clean = numerical_cols)
```

Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

```python

from sklearn.preprocessing import StandardScaler
#from sklearn.model_selection import train_test_split

scaler = StandardScaler()
X = df_without_outliers.drop(columns = "Region").values.astype(float)
y = df_without_outliers["Region"].values.astype(float)
# important to put astype(float) for code below
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
customers_scale = scaler.fit_transform(X)
```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

### Looking to the elbow we can choose 2 like the correct number of clusters

```python
wcss = []
k_range = range(1, 6) 
from sklearn.cluster import KMeans 

for i in k_range:
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init='auto', random_state=0)

    kmeans.fit(customers_scale)

    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_range, wcss, marker='o', linestyle='-')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.xticks(k_range)
plt.grid(True)
plt.show()

print("\nWCSS values for each K:")
for k, value in zip(k_range, wcss):
    print(f"K={k}: WCSS={value:.2f}")
```

```python
kmeans_2 = KMeans(n_clusters=2).fit(customers_scale)

labels = kmeans_2.predict(customers_scale)

clusters = kmeans_2.labels_.tolist()
```

```python
df_without_outliers['labels'] = clusters
```

# Challenge 5 - Data Clustering with DBSCAN

Now let's cluster the data using DBSCAN. Use `DBSCAN(eps=0.5)` to initiate the model, then fit your scaled data. In the data returned from the `.fit` method, assign the `labels_` back to `customers['labels_DBSCAN']`. Now your original data have two labels, one from K-Means and the other from DBSCAN.

```python
from sklearn.cluster import DBSCAN 

dbscan = DBSCAN(eps=2, min_samples=5)
labels = dbscan.fit_predict(customers_scale)
clusters = dbscan.labels_.tolist()
df_without_outliers['labels_DBSCAN'] = clusters
```
